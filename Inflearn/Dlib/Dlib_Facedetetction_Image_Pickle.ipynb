```python
import cv2
import face_recognition
import pickle
```

```python
# 이미지들의 특징을 train해 놓은 파일 생성(son 1~10 파일, tedy 1~10 파일) 
dataset_paths = ['D:/opencv_dnn_202005/opencv_dnn_202005/dataset/son/','D:/opencv_dnn_202005/opencv_dnn_202005/dataset/tedy/']
names = ['Son','Tedy']
number_images = 10
image_type = '.jpg'
encoding_file = 'D:/opencv_dnn_202005/opencv_dnn_202005/encodings.pickle'
# accurate와 speed 간의 trade-off, CNN is accurate but slower, HOG is faster but less accurate
model_method = 'cnn'

knownEncodings = []
knownNames = []

# 데이터 path 하나씩 돌아야함
for (i, dataset_path) in enumerate(dataset_paths) :
    name = names[i]
    
    # 이미지를 하나씩 읽어 file_name에 저장, '경로'+'숫자'+'.jpg'
    for idx in range(number_images):
        file_name = dataset_path + str(idx+1) + image_type
        
        # 이미지 읽기
        image = cv2.imread(file_name)
        print(image)
        # BGR to RGB
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # 얼굴 영역 인식
        boxes = face_recognition.face_locations(rgb, model= model_method)
        # 얼굴 영역만 encoding 하기(얼굴 영역의 특징 값들을 의미함)
        encodings = face_recognition.face_encodings(rgb, boxes)
        
        # encoding 값 저장해주기
        for encoding in encodings :
            print(file_name, name, encoding)
            knownEncodings.append(encoding)
            knownNames.append(name)
            
data = {"encodings" : knownEncodings,"names": knownNames}
f = open(encoding_file, "wb")
f.write(pickle.dumps(data))
f.close()
```
